```python
import requests

# Étape 1 : Définir la fonction pour exécuter la requête GraphQL
def fetch_generative_tokens():
    url = "https://api.v2-temp.fxhash.xyz/graphql"  # Remplacez par votre endpoint GraphQL
    query = """
    query GenerativeTokens{
      generativeTokens(take: 100) {
        displayUri
      }
    }
    """
    response = requests.post(url, json={'query': query})
    if response.status_code == 200:
        return response.json()
    else:
        raise Exception(f"Erreur lors de la requête GraphQL: {response.status_code}")

# Étape 2 : Traitement des résultats et téléchargement des images
def download_images_from_ipfs(data):
    NEXT_PUBLIC_IPFS_GATEWAY = "https://gateway.fxhash.xyz"

    for item in data['data']['generativeTokens']:
        ipfs_url = item['displayUri']
        https_url = ipfs_url.replace("ipfs://", f"{NEXT_PUBLIC_IPFS_GATEWAY}/ipfs/")

        # Télécharger l'image
        response = requests.get(https_url)
        if response.status_code == 200:
            file_name = ipfs_url.split("/")[-1] + ".png"
            with open(file_name, "wb") as f:
                f.write(response.content)
            print(f"Image {file_name} téléchargée avec succès!")
        else:
            print(f"Erreur lors du téléchargement de l'image {https_url}: {response.status_code}")

# Étape 3 : Exécution du script
if __name__ == "__main__":
    data = fetch_generative_tokens()
    download_images_from_ipfs(data)
```
